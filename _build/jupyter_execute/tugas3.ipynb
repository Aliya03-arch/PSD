2. Data sudah dideskripsikan sesuai dengan tipe data



Penjelasan visualisasi diagram batang:

- Sumbu X (horizontal) = species, berisi tiga jenis bunga Iris-setosa, Iris-versicolor, Iris-virginica.
- Sumbu Y (vertical) = nilai agregasi (Sum of id, Sum of petal_length, Sum of petal_width, Sum of sepal_length, Sum of sepal_width).
- Jadi, grafik ini memperlihatkan perbandingan jumlah (sum) petal dan sepal untuk setiap species iris.

Tipe Data:

- Numerik (Continuous/Float): petal_length, petal_width, sepal_length, sepal_width, berupa angka desimal (float).
- Kategori (Categorical): species, berupa teks (Setosa, Versicolor, Virginica).
- Integer/ID: id, biasanya berupa integer unik sebagai penanda baris data.

3. Data sudah dieksplorasi data

- Grafik Max dari setiap kolom



Grafik ini menampilkan nilai tertinggi dari setiap atribut untuk tiap spesies.

- Setosa: Petal length dan petal width tetap rendah meskipun di nilai maksimum.
- Versicolor: Nilai maksimum petal length dan petal width lebih tinggi daripada Setosa, tapi masih lebih rendah dari Virginica.
- Virginica: Mendominasi pada nilai maksimum petal length dan petal width. Menunjukkan ukuran kelopak paling besar di antara semua spesies.

Ini adalah grafik yang menjelaskan batas atas ukuran yang bisa dicapai tiap spesies.

- Grafik Min dari setiap kolom



Grafik ini menampilkan nilai terkecil dari setiap atribut untuk tiap spesies.

- Setosa: Nilai minimum petal length dan petal width sangat kecil, sesuai dengan ciri khas Setosa.
- Versicolor: Nilai minimum petal length dan petal width tetap lebih besar daripada Setosa.
- Virginica: Bahkan nilai minimum petal length dan petal width-nya lebih tinggi daripada Versicolor. Menunjukkan bahwa Virginica selalu punya kelopak lebih besar, bahkan di ukuran terkecilnya.

Ini adalah grafik yang menjelaskan batas bawah ukuran tiap spesies.

- Grafik Average dari setiap kolom



Grafik ini menunjukkan nilai rata-rata (mean) dari setiap atribut bunga untuk setiap spesies.

- Setosa: Rata-rata sepal width lebih besar dibanding dua spesies lain. Rata-rata petal length dan petal width paling kecil.
- Versicolor: Rata-rata petal length dan petal width sedang (antara Setosa dan Virginica). Sepal length lebih panjang dari Setosa, tapi lebih pendek dari Virginica.
- Virginica: Memiliki rata-rata sepal length paling panjang. Rata-rata petal length dan petal width paling besar.

Ini adalah grafik yang membantu mengenali ciri khas umum tiap spesies dengan melihat nilai tengahnya.

- Grafik Jumlah Data per Species (Kelas)



Grafik batang ini menampilkan banyaknya data (jumlah baris) untuk tiap kelas bunga Iris:
- Iris Setosa: jumlah datanya 50
- Iris Versicolor: jumlah datanya 50
- Iris Virginica  jumlah datanya 50

Jadi setiap batang di grafik akan memiliki tinggi yang sama, karena dataset Iris seimbang: masing-masing spesies memiliki jumlah data yang sama (50).

Penjelasan Lengkap:
a. Missing Value



Grafik batang ini menampilakan Missing Values (kolom Missing_ID, Missing_PetalLength, dll).

Dari visual:
- Mayoritas nilai = 0, artinya data lengkap (tidak ada banyak missing value).
- Kalau ada bar kecil selain 0, berarti ada data hilang untuk kolom tertentu.

Kesimpulannya, dataset ini hampir tidak ada missing value, jadi tidak perlu banyak imputasi (mengisi data kosong).

b. Imbalance (Penyeimbangan Data)



Grafik batang ini menampilakan data imbalance
- Sumbu X = kategori species (Iris-setosa, Iris-versicolor, Iris-virginica).
- Sumbu Y = jumlah data (CountPerSpecies).
- Dataset Iris memiliki 150 record total  masing-masing species (setosa, versicolor, virginica) berjumlah 50 data.
- Jadi distribusinya :
    - setosa: 50
    - versicolor: 50
    - virginica: 50

Hasil grafik menunjukkan ketiga bar tingginya sama besar, artinya masing-masing species punya jumlah data yang seimbang.

c. Outlier (Data Menyimpang)



Scatter plot dipakai untuk mendeteksi outlier.
- Kiri: sumbu X = petal_length, sumbu Y = petal_width.
- Terlihat pola linear (makin panjang petal, makin lebar).
- Outlier terlihat sebagai titik jauh di bawah cluster utama (misalnya petal_length kecil tapi petal_width tidak seimbang).
- Kanan: sumbu X = sepal_length, sumbu Y = sepal_width.
- Ada pemisahan kelompok sesuai warna species (Iris-setosa, Iris-versicolor, Iris-virginica).
- Outlier = titik yang jauh menyendiri dari kelompok dominan.

Dari scatter ini, tidak terlalu banyak outlier, artinya dataset Iris cukup bersih, tapi ada beberapa titik di bagian bawah yang bisa dianggap kandidat outlier.
